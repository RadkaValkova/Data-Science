```python
text = ''
with open('1342-0.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
words_in_text = re.split("\W+",text.lower())
```

```python
word_counter = Counter(words_in_text)
pride_counter = word_counter['pride']
```

```python
pride_counter
```

```python
predjudice_counter = word_counter['prejudice']
```

```python
predjudice_counter
```

```python
stop = set(stopwords.words('english'))
```

```python
cleaned_text = Counter([w for w in words_in_text if w not in stop])
```

```python
cleaned_text.most_common(20)
```
