```
1. CRIM      per capita crime rate by town
2. ZN        proportion of residential land zoned for lots over 
             25,000 sq.ft.
3. INDUS     proportion of non-retail business acres per town
4. CHAS      Charles River dummy variable (= 1 if tract bounds 
             river; 0 otherwise)
5. NOX       nitric oxides concentration (parts per 10 million)
6. RM        average number of rooms per dwelling
7. AGE       proportion of owner-occupied units built prior to 1940
8. DIS       weighted distances to five Boston employment centres
9. RAD       index of accessibility to radial highways
10. TAX      full-value property-tax rate per $10,000
11. PTRATIO  pupil-teacher ratio by town
12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
             by town
13. LSTAT    % lower status of the population
14. MEDV     Median value of owner-occupied homes in $1000's
```

````python
plt.hist(housing_prices['crim'], bins=20) # Извикваме хистограмата

plt.title('Distribution of crime rates')

plt.xlabel('Crime rate [%]') # добавяме надписи по осите - задължително!!
plt.ylabel('count')

plt.xlim(0,100) # стойностите по оста се разпридилят от 0 до 100

plt.show() # ичиства допълнителната информация, която връща jupiter 
````

````python
plt.hist(housing_prices['crim'], bins=20, cumulative=True )

plt.title('Cummulative Distribution of crime rates')

plt.xlabel('Crime rate [%]')
plt.ylabel('count')

plt.xlim(0,100)

plt.show() 
````

````python
plt.hist(housing_prices['crim'], bins=20, cumulative=True, histtype='step' )

plt.title('Cummulative Distribution of crime rates')

plt.xlabel('Crime rate [%]')
plt.ylabel('count')

plt.xlim(0,100)

plt.show() 
````

````python
plt.boxplot(housing_prices['crim'])

plt.xticks([1], ['Crime rate'])
plt.ylabel('Count')

plt.show()
````

````python
housing_prices['rad'].unique()
````

````python
num_houses_by_accessability = housing_prices.groupby('rad').size()
````

````python
num_houses_by_accessability
````

````python
plt.bar(num_houses_by_accessability.index, num_houses_by_accessability)
plt.xticks(list(range(1,9)) + [24])
plt.show()
````

````python
prices_by_accessability = housing_prices.groupby('rad')['medv'].median()
````

````python
prices_by_accessability
````

````python
plt.bar(prices_by_accessability.index, prices_by_accessability)

plt.xticks(list(range(1,9)) + [24])
plt.xlabel('Index of accessability')
plt.ylabel('Median of median prices [x $1000]')
plt.show()
````

````python
plt.scatter(housing_prices['rm'], housing_prices['medv'], s=5)

plt.xlabel('Number of rooms')
plt.ylabel('Median values [x $1000]')
plt.show()
````

````python
indices_of_outliers = housing_prices[(housing_prices.rm <6) & (housing_prices.medv < 20)].index # намираме индексите им и ги присвояваме
````

Премахваме аутлайерите от дейтасета, но новите данни запаметяваме в нова променлива като така запазваме старите данни непроменени

````python
housing_prices_no_outliers = housing_prices.drop(indices_of_outliers) 
````

````python
housing_prices_no_outliers
````

Правим нова графика вече с почистените от аутлайери данни
